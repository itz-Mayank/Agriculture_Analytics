### `Top 5 Districts by Wheat Yield Increase Over the Last 5 Years`

```python
# Top 5 Districts by Wheat Yield Increase Over the Last 5 Years
cursor.execute('''
    SELECT 
        recent.Dist_Name,
        recent.WHEAT_YIELD AS Latest_Yield,
        past.WHEAT_YIELD AS Yield_5_Years_Ago,
        (recent.WHEAT_YIELD - past.WHEAT_YIELD) AS Yield_Increase
    FROM 
        agri_data AS recent
    JOIN 
        agri_data AS past
    ON 
        recent.Dist_Name = past.Dist_Name
    WHERE 
        recent.Year = (SELECT MAX(Year) FROM agri_data)
        AND past.Year = (SELECT MAX(Year) FROM agri_data) - 5
    ORDER BY 
        Yield_Increase DESC
    LIMIT 5;
''')

top_districts_join = cursor.fetchall()
print("Top 5 Districts (JOIN method):")
for row in top_districts_join:
    print(f"District: {row[0]}, Yield Increase: {row[3]:.2f}")
```

### `States with the Highest Growth in Oilseed Production (5-Year Growth Rate)`

```python
# States with the Highest Growth in Oilseed Production (5-Year Growth Rate)

cursor.execute('''
    SELECT 
        recent.State_Name,
        recent_prod AS Latest_Production,
        past_prod AS Production_5_Years_Ago,
        ROUND(((recent_prod - past_prod) / past_prod) * 100, 2) AS Growth_Percent
    FROM (
        SELECT State_Name, SUM(OILSEEDS_PRODUCTION) AS recent_prod
        FROM agri_data
        WHERE Year = (SELECT MAX(Year) FROM agri_data)
        GROUP BY State_Name
    ) AS recent
    JOIN (
        SELECT State_Name, SUM(OILSEEDS_PRODUCTION) AS past_prod
        FROM agri_data
        WHERE Year = (SELECT MAX(Year) FROM agri_data) - 5
        GROUP BY State_Name
    ) AS past
    ON recent.State_Name = past.State_Name
    WHERE past_prod > 0
    ORDER BY Growth_Percent DESC
    LIMIT 5;
''')

top_states_growth = cursor.fetchall()

print("Top 5 States by Oilseed Production Growth Rate (Last 5 Years):")
for row in top_states_growth:
    print(f"State: {row[0]}, Growth: {row[3]}%, From {row[2]} to {row[1]}")


```

### `District-wise Correlation Between Area and Production for Major Crops (Rice, Wheat, and Maize)`

```python
# District-wise Correlation Between Area and Production for Major Crops (Rice, Wheat, and Maize)
import pandas as pd

crops = ['RICE', 'WHEAT', 'MAIZE']
results = []

for crop in crops:
    area_col = f'{crop}_AREA'
    prod_col = f'{crop}_PRODUCTION'

    for dist, group in df.groupby('Dist_Name'):
        if len(group[area_col].dropna()) > 1:
            corr = group[area_col].corr(group[prod_col])
            if pd.notnull(corr):
                results.append({
                    'District': dist,
                    'Crop': crop,
                    'Correlation': round(corr, 3)
                })

correlation_df = pd.DataFrame(results)
print(correlation_df.sort_values(by=['Crop', 'Correlation'], ascending=[True, False]))

```

### `Yearly Production Growth of Cotton in Top 5 Cotton Producing States`

```python
# Yearly Production Growth of Cotton in Top 5 Cotton Producing States
# Step 1: Get top 5 states by total cotton production
cursor.execute('''
    SELECT State_Name
    FROM agri_data
    GROUP BY State_Name
    ORDER BY SUM(COTTON_PRODUCTION) DESC
    LIMIT 5;
''')

top_states = [row[0] for row in cursor.fetchall()]

# Step 2: Get year-wise cotton production for these states
# Create placeholder string for SQL parameters
placeholders = ', '.join(['%s'] * len(top_states))

query = f'''
    SELECT Year, State_Name, SUM(COTTON_PRODUCTION) AS Total_Production
    FROM agri_data
    WHERE State_Name IN ({placeholders})
    GROUP BY Year, State_Name
    ORDER BY State_Name, Year;
'''

cursor.execute(query, top_states)
rows = cursor.fetchall()

# Step 3: Load into DataFrame for analysis
import pandas as pd

df = pd.DataFrame(rows, columns=['Year', 'State_Name', 'Total_Production'])

# Step 4: Calculate year-over-year growth (optional)
df['Production_Growth'] = df.groupby('State_Name')['Total_Production'].pct_change() * 100
df['Production_Growth'] = df['Production_Growth'].round(2)

df
```

### `Annual Average Maize Yield Across All States`

```python
# Annual Average Maize Yield Across All States
cursor.execute('''
    SELECT 
        Year, 
        ROUND(AVG(MAIZE_YIELD), 2) AS Avg_Maize_Yield
    FROM 
        agri_data
    WHERE 
        MAIZE_YIELD IS NOT NULL
    GROUP BY 
        Year
    ORDER BY 
        Year;
''')

maize_yield = cursor.fetchall()

print("Annual Average Maize Yield Across All States:")
for row in maize_yield:
    print(f"Year: {row[0]}, Average Yield: {row[1]} kg/ha")

```
